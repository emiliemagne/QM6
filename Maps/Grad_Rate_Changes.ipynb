```python
import pandas

import matplotlib as plt
%matplotlib inline

graduation_rates_19 = pandas.read_csv('./drive/MyDrive/Johann/UCL/Second Year/Quantitative Methods 2/FinalProject/GRAD_RATE_AND_OUTCOMES_2019.csv')
graduation_rates_20 = pandas.read_csv('./drive/MyDrive/Johann/UCL/Second Year/Quantitative Methods 2/FinalProject/GRAD_RATE_AND_OUTCOMES_2020.csv')
```

```python
graduation_rates_19 = graduation_rates_19.loc[graduation_rates_19['aggregation_type'] == 'County']
graduation_rates_19 = graduation_rates_19.loc[graduation_rates_19['membership_desc'] == '2015 Total Cohort - 4 Year Outcome']
graduation_rates_19 = graduation_rates_19.loc[graduation_rates_19['subgroup_name'] == 'All Students']
graduation_rates_19
```

```python
graduation_rates_20 = graduation_rates_20.loc[graduation_rates_20['aggregation_type'] == 'County']
graduation_rates_20 = graduation_rates_20.loc[graduation_rates_20['membership_desc'] == '2016 Total Cohort - 4 Year Outcome']
graduation_rates_20 = graduation_rates_20.loc[graduation_rates_20['subgroup_name'] == 'All Students']
graduation_rates_20
```

```python
graduation_rates_19 = graduation_rates_19.replace('%', '', regex=True)
graduation_rates_20 = graduation_rates_20.replace('%', '', regex=True)
graduation_rates_20 = graduation_rates_20.replace('County: ', '', regex=True)
graduation_rates_19 = graduation_rates_19.replace('County: ', '', regex=True)
```

```python
from pandas.core.frame import DataFrame
difference = []
for i in range(0, len(graduation_rates_20['aggregation_name'])):
  difference_ = graduation_rates_20['grad_pct'].astype(float).iloc[i] - graduation_rates_19['grad_pct'].astype(float).iloc[i]
  difference.append(difference_ * -1)
print(difference)
comparison = pandas.DataFrame()
comparison['County'] = graduation_rates_20['aggregation_name']
comparison['Difference'] = difference
data_path = './drive/MyDrive/Johann/UCL/Second Year/Quantitative Methods 2/FinalProject/Graduation_Rate_differences.csv'
comparison.to_csv(data_path)
comparison = comparison.sort_values(by = ['Difference'])
comparison_negative = comparison[56:]
comparison_positive = comparison[0:56]
comparison_negative
comparison
```

```python
AP_differences_negative = pandas.read_csv('./drive/MyDrive/Johann/UCL/Second Year/Quantitative Methods 2/FinalProject/AP_differences_negative.csv')
ambos = comparison_negative.merge(AP_differences_negative, left_on = 'County', right_on = 'County', how = 'inner')
ambos
#only cattaraugus and washington are missing. 
#it can, however, be clearly seen that only 6 counties (cattaraugus, tompkins, washington, essex, yates, and hamilton) had a negative change in graduation rates
#while 18 counties had a negative change in AP scores, 10 of which were larger than 0.1 points but only 2 of which were larger than 1 (Allegany and Yates). 
#what do these counties that performed badly have in common. 
```
